```python
import pandas as pd
import os

# 读取数据
df = pd.read_csv('./turbine_prediction_selected_features_400turbines.csv')

# 检查是否存在 cluster 列
if 'cluster' not in df.columns:
    raise ValueError("该 CSV 文件中不包含 'cluster' 列。")

# 创建输出文件夹
output_dir = "cluster_split"
os.makedirs(output_dir, exist_ok=True)

# 按 cluster 分组并保存为不同文件
for cluster_id, cluster_df in df.groupby('cluster'):
    filename = f"{output_dir}/cluster_{cluster_id}.csv"
    cluster_df.to_csv(filename, index=False)
    print(f"saved: {filename}")
```
