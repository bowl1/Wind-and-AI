```python
import pandas as pd

# === input file ===
feature_file = "./selected feature statistics/features_in_group_stage_200_turbines.csv"
cluster_file = "./group_stage_results/first_group/best_n8_k5_r9_assignment.csv" 
output_file = "features_cluster_0_1st.csv"  
    
features_df = pd.read_csv(feature_file)
clusters_df = pd.read_csv(cluster_file)

# === merge data (by GSRN) ===
merged_df = pd.merge(features_df, clusters_df, on="GSRN", how="inner")

# === extract data where cluster is 0 or 1 ===
subset_df = merged_df[merged_df["cluster"].isin([1])]

subset_df.to_csv(output_file, index=False)

print(f"complete, total {len(subset_df)} rows of data, saved as: {output_file}")
```
