## the first time group count:

```python
import pandas as pd

input_file = "./group_stage_results_new/1st/best_n7_k3_r7_assignment.csv"

df = pd.read_csv(input_file)

# count turbines in each cluster
cluster_counts = df["cluster"].value_counts().sort_index()
print("count of turbines in each cluster:")
for cluster_id, count in cluster_counts.items():
    print(f"Cluster {cluster_id}: {count} turbines")
```

```python

# === input file ===
feature_file = "./selected feature statistics/features_in_group_stage_50_turbines.csv"
cluster_file = "./group_stage_results_new/1st/best_n7_k3_r7_assignment.csv" 
output_file = "features_cluster_0_1st.csv"  
    
features_df = pd.read_csv(feature_file)
clusters_df = pd.read_csv(cluster_file)

# === merge data (by GSRN) ===
merged_df = pd.merge(features_df, clusters_df, on="GSRN", how="inner")

# === extract data where cluster is 0 or 1 ===
subset_df = merged_df[merged_df["cluster"].isin([0])]

subset_df.to_csv(output_file, index=False)

print(f"complete, total {len(subset_df)} rows of data, saved as: {output_file}")
```
