```python


# ËÆæÁΩÆË∑ØÂæÑ
base_paths = {
    "masterdata": "energinet/masterdatawind.parquet",
    "observations": "energinet/observations",
    "settlement": "energinet/settlement",
    "settlement2": "energinet/settlement2/Year=2016"
}

# ËØªÂèñ masterdata
print("\nMASTERDATA")
try:
    df = pd.read_parquet(base_paths["masterdata"])
    display(df.head())  # Jupyter ÊîØÊåÅÊõ¥ÁæéËßÇÊòæÁ§∫
except Exception as e:
    print(f"Error reading masterdata: {e}")

# ËØªÂèñ observations ‰∏ãÁöÑÊØè‰∏™Êñá‰ª∂
print("\n OBSERVATIONS")
obs_dir = base_paths["observations"]
for file in os.listdir(obs_dir):
    if file.endswith(".parquet"):
        print(f"\n {file}")
        try:
            df = pd.read_parquet(os.path.join(obs_dir, file))
            display(df.head())
        except Exception as e:
            print(f" Error reading {file}: {e}")

# ËØªÂèñ settlement Êüê‰∫õÂπ¥‰ªΩ
print("\n SETTLEMENT (sample years)")
settle_dir = base_paths["settlement"]
for file in sorted(os.listdir(settle_dir))[:3]:  # ÈôêÂà∂Ââç3‰∏™
    if file.endswith(".parquet"):
        print(f"\n {file}")
        try:
            df = pd.read_parquet(os.path.join(settle_dir, file))
            display(df.head())
        except Exception as e:
            print(f"Error reading {file}: {e}")

# ËØªÂèñ settlement2 ÊüêÂá†‰∏™ part Êñá‰ª∂
print("\n SETTLEMENT2 (sample parts)")
settle2_dir = base_paths["settlement2"]
for file in sorted(os.listdir(settle2_dir))[:3]:
    if file.endswith(".parquet"):
        print(f"\n {file}")
        try:
            df = pd.read_parquet(os.path.join(settle2_dir, file))
            display(df.head())
        except Exception as e:
            print(f" Error reading {file}: {e}")
```
